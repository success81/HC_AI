```python
import os
import vertexai
from vertexai.preview.generative_models import GenerativeModel

# Set up authentication and Vertex AI environment
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "key.json"
PROJECT_ID = "winter-cogency-436501-g9"
REGION = "us-central1"

vertexai.init(project=PROJECT_ID, location=REGION)

# Simple function to send a question to the model and get a response
def call_ai_model(question):
    generative_model = GenerativeModel(model_name="gemini-1.5-pro")
    
    # Prepare the prompt
    prompt = f"Answer this question: {question}"

    # Generate response
    response = generative_model.generate_content(
        contents=[{"role": "user", "parts": [{"text": prompt}]}],
        generation_config={"temperature": 0.5, "max_output_tokens": 100}
    )
    
    return response.text.strip()

# Example usage
question = "What color is the sky?"
print("Response:", call_ai_model(question))
```
